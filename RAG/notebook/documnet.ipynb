### Use Directory Loader to Load and Create Structural Documents

```python
# Loader to Load and Create Structural Documents of All Files in a Directory Simultaneously
from langchain_community.document_loaders import DirectoryLoader

# Load all txt files in directory
dir_loader = DirectoryLoader(
    "../data/text_files",
    glob = "**/*.txt", # Patern to match all files
    loader_cls = TextLoader, # Loader class to use
    loader_kwargs = {"encoding":"utf-8"},
    show_progress = False
)

txt_docs = dir_loader.load()
txt_docs
```

```python
# Load all pdf files in directory
pdf_dir_loader = DirectoryLoader(
    "../data/pdf_files",
    glob = "**/*.pdf", # Patern to match all files
    loader_cls = PyMuPDFLoader, # Loader class to use
    show_progress = False
)

pdf_docs = pdf_dir_loader.load()
pdf_docs
```
